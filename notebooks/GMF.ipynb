```python
# Extract and  historical financial data using YFinance for TSLA

import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Define the ticker symbol
tickerSymbol = 'TSLA'

# Get data on this ticker
tickerData = yf.Ticker(tickerSymbol)

# Get the historical prices for this ticker
tickerDfTS = tickerData.history(period='1d', start='2010-1-1', end='2021-1-1')

# Plot the close prices
plt.figure(figsize=(12,6))
plt.plot(tickerDfTS['Close'])
plt.title('Close Price of ' + tickerSymbol)
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

```python
# Normalize or scale the data if required, especially for machine learning models.

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

# Normalize the data
scaled_data = scaler.fit_transform(tickerDfTS['Close'].values.reshape(-1, 1))

# Create a new DataFrame
dfts = pd.DataFrame(scaled_data, columns=['Close'])
print(dfts.head())
```
